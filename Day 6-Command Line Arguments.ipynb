# Command Line Arguments

>Command line arguments are the values pased to the pytho program at comman prompt

arguments are entered from keyboard seperated by a space

>These arguments are stored by default in the from ofstrings in a list named as 'argv' which is available in sys module



program file name is also included in the command line argument list.

```python
#importing sys module
import sys
```

In the following example we have taken kamal Singh as a command line argument but we observe that compiler is considering them as two different arguments showing number of arguments as 3



But if we take the name as " 'Kamal Singh' "  or  ' "Kamal Singh" ' it is considered as one single argument 



```python
#cmd.py

#program to extract the command line arguments and printing them
import sys

n=len(sys.argv) #returns number of command line arguments
args=sys.argv
print('Number of Command line arguments are: %i'%n)
print(args)
for a in args:
    print(a)
```

```python
#arithematic.py

#taking two command line arguments and performing arithematic operations
import sys

n=len(sys.argv)
n1=int(sys.argv[1])
n2=int(sys.argv[2])
print('Arithematic operations: ')
print('Sum={0} \nDifference: {1} \nProduct: {2}\nDivision: {3}'.format((n1+n2),(n1-n2),(n1*n2),(n1/n2)))

```

```python
#even_odd_sum.py

#finding odd and even elements sum for elements provided in command line
import sys
sum_eve=sum_odd=0
lst=sys.argv[1:]
print(lst)
print(len(lst))
for a in lst:
    x=int(a)
    if x%2==0:
        sum_eve+=x
    else:
         sum_odd+=x
print('sum of even elements: ',sum_eve)
print('sum of odd elements: ',sum_odd)
```

```python
#eve_odd_sum_with_list.py

#finding odd and even elements sum for elements of a list
import sys
sum_eve=sum_odd=0
samp1=sys.argv[1].replace('[','')
samp2=samp1.replace(']','')

'''print(samp2)
print(type(samp2))
print(len(samp2))
print(type(lst))'''
lst=list(samp2.split(","))
print('Ur given elements are: ',lst)
for a in lst:
    x=int(a)
    if x%2==0:
        sum_eve+=x
    else:
         sum_odd+=x
print('sum of even elements: ',sum_eve)
print('sum of odd elements: ',sum_odd)
```
